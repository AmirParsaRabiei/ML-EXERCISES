reconstruct an image using SVD PCA

```python
import cv2
from numpy import linalg as LA
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('photo_2022-12-20_07-35-10.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

U , S , V = LA.svd(gray)
print(U.shape)
print(S.shape)
print(V.shape)
print(S)

# cv2.imshow("photo_2022-12-20_07-35-10.jpg",gray)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# for i in [2]:
for i in [5,10,15,20,30,50,100]:
    reconstructed_img = np.matrix(U[:, :i]) * np.diag(S[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstructed_img, cmap='gray')
    plt.title("n={}".format(i))
    plt.show() 
```
