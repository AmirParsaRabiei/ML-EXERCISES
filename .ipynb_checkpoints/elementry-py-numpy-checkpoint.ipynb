```python
# creating array
a = np.array([1,2,3])
print(a)
print('='*80)
# ones matrix
a = np.ones(3)
print(a)
print('='*80)
# zeors
a = np.zeros(3)
print(a)
print('='*80)
# generates random floating-point numbers uniformly distributed between 0 and 1 (exclusive of 1)
a = np.random.random()
print(a)
print('='*80)

a = np.random.random(size=5)
print(a)
print('='*80)

# 2d
a = np.random.random(size=(5,5))
print(a)
print(type(a)) # ndarray : n-dimensional array (array haye chand boudi) ke be ina migan tensor
print('='*80)

# to list mamoli mishe inkaro kard mesle addad behesh negah mikone
a = [1,2,3] + [2,3,4,5,6]
print(a)
print('='*80)

# vali to np.array chon tensor hast, hatman bayad vectorize behesh negah koni va len yeksan bashe
a = np.array([1,2,3])
b = np.array([1,1,1])
print(a+b)
print('='*80)
a = np.array([1,2,3])
b = np.ones(3)
print(a+b)
print('='*80)


"""
dar halati ke ba addad ha hastim nmishe [1,2,3]+10 kard...
vali ba np array ha mishe
"""
# broadcasting
a = np.array([1,2,3]) + 2
print(a)
print('='*80)

"""
kolan hamon kar haye ke mikonim element be element(elementvize) mesle matrix(tensor)
anjam mide
msln to ML bias ro ba multiply kardan ye nouron to 0.7 mitonam kole
element hasho * 0.7 konam...
"""

# Aggregation
a = np.array([1,2,3])
print(np.max(a))
print(np.min(a))
print(np.sum(a))
print(np.prod(a)) # multiplying all the elements together
print(np.std(a)) # standard deviation
print(np.mean(a)) # average
print(math.sqrt(1))
```

```python
import numpy as np

def statistics(nparray):
    a = np.mean(nparray)
    b = np.std(nparray)
    sumation = np.sum(nparray)
    return a,b,sumation

tensor = np.array([44,5,6,87,5,56,20])

mean, deviation, sumation = statistics(tensor)
print(f'mean is {mean} / deviation is {deviation} / sumation is {sumation}')
print('='*20)
"""
data type haye tensor bayad hamgen bashan vagarna kar nmikone...
msln age to tensor string bezaram, data type kole tensor string mishe
"""
print(tensor.dtype)

print('='*20)
# bezor ham mitonam data type ro malom konam
a = np.array([44,5,6,87,5,56,20], dtype='float32')
print(a.dtype)
print(a.shape)
print(type(a.shape))
print(a.shape[0])
print(a.ndim)
print('='*20)

# arithmetic(+-/*) in numpy(with matrix)
a = np.array([[1,2],[3,4],[5,6]]) # 3 row , 2 column (2d)
b = np.array([[4,5],[6,5], [5,5],[5,5]]) # 4 row and 2 column (2d) 
c = np.array([1,2,3,4,5,6,7]) # 1 row and 7 column (1d)
d = np.array([[[12,12,8],[1,2,8]],[[56,59,8],[4,9,8]],[[5,5,8],[9,8,8]],[[4,5,8],[48,48,8]]])
print(a.shape) 
print(a + 5)
print('='*20)
print(d.shape)
print(d)

  
```

```python
"""
broadcast faghat vaghti mishe ke ya row 1 va column equal to column digari
ya row equal row digari va column 1
va halate badi ham ine ke shape yeksan bashe 
"""

a = np.random.randint(0, 10, size=(2,3))
b = np.random.randint(0, 10, size=(1,3))
c = np.random.randint(0, 10, size=(2,1))
d = np.random.randint(0, 10, size=(3,4))
print(a)
print('='*10)
print(b)
print('='*10)
print(c)
print('='*10)
print(a+b) # broadcasting
print('='*10)
print(a+c) # broadcasting

print('='*80)

"""
vaghti migi a*b miad element-wise (element be element) zarb mikone
vai age bekhay dghighan mesle matrix mathematic zarb kone
bayad az dot use koni 
va inke shape ham bayad intor bashe ke n*m m*k => n*k ta zarb anjam beshe
"""
z = np.dot(a,d)
print(z)
print('='*80)
print(z.shape)
print('='*40)

"""
slicing : msln mikham tensor a faghat behem column 1 va 2 ro bede(indexing az 0 hast)
"""
print(a[:,1:3])

print('='*80)

# matrix aggregation
print(a)
print('='*80)
print(a.max())
print('='*80)
print(a.max(axis=1))
print('='*80)
print(a.max(axis=0))
print('='*80)
```
