```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data = np.random.normal(0, 1, 1000)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the histogram
ax.hist(data, bins=20, edgecolor='black', alpha=0.7)

# Set the title and labels
ax.set_title('Histogram of Random Data', fontsize=16)
ax.set_xlabel('Value', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)

# Set the tick labels font size
ax.tick_params(axis='both', labelsize=12)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add grid lines
ax.grid(axis='y', alpha=0.7)

# Adjust the layout
fig.tight_layout()

# Display the plot
plt.show()
```

```python
"""
ye chiz bahale dig az tuple unpacking mikham begam
ba star, moghe unpack kardan, dobare miad ye tuple dig ham pass mide...
ba mesal neshon midam
"""

# without star unpacking
a, b , c = 15 , 18 , 20
print(f'a = {a}, b= {b}, c= {c}')

print('='*20)
# with star (Only one unpack operation allowd in list)
*a, b , c = 15 , 18 , 20 , 40 , 50 , 60 , 70 , 80 , 90 , 100
print(f'a = {a}, b= {b}, c= {c}')
print(f'a type ={type(a)}')

print('='*20)
# with star
a, *b , c = 15 , 18 , 20 , 40 , 50 , 60 , 70 , 80 , 90 , 100
print(f'a = {a}, b= {b}, c= {c}')
print(f'b type ={type(b)}')

print('='*20)
# with star
a, b , *c = 15 , 18 , 20 , 40 , 50 , 60 , 70 , 80 , 90 , 100
print(f'a = {a}, b= {b}, c= {c}')
print(f'c type ={type(c)}')
```

```python
"""
mishe input haro ba * gozashtan, be sorat ye list pass kard be function
va yani chanding argument mishe dad be function
darvaghe tuple mikone va pass mide(hamon tuple unpacking)

dar vaghe vaghti nimdoni ye input ya output chandta argument dare azash
use mikoni

"""

# input star
def sum_numbers(*numbers):
    print(type(numbers))
    total = 0
    for number in numbers:
        total += number
    return total

resualt = sum_numbers(1,2,3,4,5)
print(resualt)
print(type(resualt))

print('='*20)
# age nakham tuple beshe, ya bayad badan tabdil konam ya inke mostaghim list bedam
def sum_numbers(numbers):
    total = 0
    for number in numbers:
        total += number
    return total

resualt = sum_numbers([1,2,3,4,5])
print(resualt)
print(type(resualt))

print('='*20)
# without star (it't bug)
def sum_numbers(numbers):
    total = 0
    for number in numbers:
        total += number
    return total

resualt = sum_numbers(1,2,3,4,5)
print(resualt)
print(type(resualt))

```
