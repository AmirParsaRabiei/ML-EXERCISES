```python
"""
سوال 1(
یک نوت بوک پایتونی درست کنید و در آن یک list دلخواه شامل اعداد float، int و همچنین str درست کنید.
بر روی آن for بزنید و تمام عناصر عددی و بزگتر از 20 آن را چاپ کنید. 
"""

#1stmethod | type-checking

my_list = [5, 5.5, 4, 3, 50, 550.2, 22, 1, -1, 'test']

for item in my_list:
    if isinstance(item, (int, float)) and item > 20:
        print(item)

print('='*20)

#2nd-method | try-except

my_list = [5, 5.5, 4, 3, 50, 550.2, 22, 1, -1, 'test']

for item in my_list:
    try:
        if item > 20:
            print(item)
    except TypeError:
        print(f"Warning: '{item}' is not a numeric value and was skipped.")

print('='*20)

#3nd | Filter Out Non-Numeric Elements

my_list = [5, 5.5, 4, 3, 50, 550.2, 22, 1, -1, 'test']

numeric_list = list(filter(lambda x: isinstance(x, (int, float)), my_list))

for item in numeric_list:
    if item > 20:
        print(item)

print('*'*20)


#4nd | using type condition
for item in my_list:
    if type(item) in (int, float) and item > 20:
        print(item)

#5nd | another notation
for item in my_list:
    if type(item)== int or type(item)==float and item > 20:
        print(item)
```

```python
"""
سوال 2(
یک دیکشنری درست کنید.
در یک for از کاربر اسم و نمره بگیری د و این دیکشنر ی را مقداردهی کنید، طوری که کلیدها نام و مقادیر معدل باشد.
در آخر معدل این افراد و میانگین کل کلاس را چاپ کنید
"""

scores = {}

while True:
    name = input("Enter the name (or press Enter to finish): ")
    if name == "":
        break
    
    score = int(input(f"Enter the score for {name}: "))
    scores[name] = score

print("Scores:")
for name, score in scores.items():
    print(f"{name}: {score}")

# 1st method | python
print('-'*20)
average = sum(scores.values()) / len(scores.values())
print(f'Average: {average}')

# 2nd method | numpy

import numpy as np

# list mikonim chon scores.values() dict type hast va np array ba list kar mikone
np_average = np.mean(list(scores.values()))

print(f"Avg is:{np_average}")
```

```python
"""
سوال 5(
یک تابع به نام five_np در پایتون تعریف کنید که ورودی اش یک عدد باشد و خروجی اش یک ماتریس پر از 5 باشد.
برا ی مثال با فراخوانی five_np(2 )یک ماتریس نامپای 2 در 2 که در همه خانه های آن 5 هست برگرداند.
"""

import numpy as np

# 1st | np full
def five_np(num):
    return np.full((num, num), 5)
print(five_np(2))

print('='*20)
# 2nd | np ones
def five_np_ones(num):
    return np.ones((num,num))*5
print(five_np_ones(4))
```

```python
"""
سوال 6(
 یک وکتور)آرایه یک بعدی( پایتون به صورت رندم و تصادفی با طول 500.000 بسازید. 
 و با پیاده ساز ی به دو روش زیر زمان اجرا را مقایسه و چاپ کنید
 الف( یک تابع پایتونی بنویسید که با for زدن روی این آرایه بزرگترین عدد را برگرداند.
ب( با توابع خود numpy که میدانی م پیاده ساز ی vectorize انجام داده اند ماکزیمم این لیست را حساب کنید.
"""
import numpy as np

arr = np.random.rand(500000)
print(arr)

# A
def max_num(arr):
    max_num = arr[0]
    for num in arr:
        if num > max_num:
            max_num = num
    return max_num
print(max_num(arr))

# B
print(np.max(arr))
```

```python
"""
سوال 7(
 یک عکس دلخواه لود کنید، تمام خانه های آن را به عالوه 10 کنید و ذخیره کنید. عکس نهایی و عکس اولیه را در پوشه تمرین بزارید و
ارسال کنید
"""

"""
download image with bash
!wget image_adress
"""

# 1st method | np.clip

from PIL import Image
import numpy as np
import os

def load_and_modify_image(image_path, output_folder):
    # Load the image
    img = Image.open(image_path)
    img = img.convert('RGB')  # Ensure image is in RGB format

    # Convert image to numpy array
    data = np.array(img)

    # Convert data type to int32 to handle addition without overflow
    data = data.astype('int32')

    # Add 50 to each pixel. Since pixel values must be in [0,255], we clamp them using np.clip
    modified_data = np.clip(data + 50, 0, 255)
    # inam mishod nevesht jaye clip:

    # Add 50 to each pixel
    # modified_data = data + 50

    # Identify pixels that exceed 255 and set them to 255
    # modified_data[modified_data > 255] = 255
    
    # Convert data type back to uint8
    modified_data = modified_data.astype('uint8')

    # Convert numpy array back to image
    modified_img = Image.fromarray(modified_data, 'RGB')

    # checking the maximum value
    max_value = np.max(np.array(modified_data))

    # Define file paths
    original_image_path = os.path.join(output_folder, 'original_image1.png')
    modified_image_path = os.path.join(output_folder, 'modified_image1.png')

    # Save the original image
    img.save(original_image_path)

    # Save the modified image
    modified_img.save(modified_image_path)

    print(f"Images saved in {output_folder}")
    print("Original image path:", original_image_path)
    print("Modified image path:", modified_image_path)
    print(f"Max_value:{max_value}")

# Specify the image path and output folder
image_path = 'photo_2022-12-20_07-35-10.jpg'  # Update this path to your image's path
output_folder = 'Q7_image'  # Define or update output directory

# Create output directory if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Call the function
load_and_modify_image(image_path, output_folder)


print('================================================')


# 2nd method | just adding to plt.imgshow(img +10)

import matplotlib.pyplot as plt

img = plt.imread('photo_2022-12-20_07-35-10.jpg')
print(type(img))
print(img.dtype)
# code zir rang haro kharab mikone vali code 
plt.imshow(img+50)
plt.title('Image with pixel values increased by 50(overflow)')
plt.show()

# 3nd method | add astype and cliping

img2 = plt.imread('photo_2022-12-20_07-35-10.jpg')

""" chon unsigned int8 darim yani unsigned integer 1 byte darim ke range decimal [0,255] hast va 
vaghti ma pixel add mizanim overflow anjam mishe va rang ha kharab mishe
"""
print(f"image data type: img2.dtype")
# int 32 mikonam ke ba increase 50 pixeli, value ham ja beshe bedone overflow(ke taghir rang bede)
img2 = img2.astype('int32') + 50

# img2[img2>255] miad har pixeli bozorg tar az 255 true barmigardone va badesh entesab midam behesh true haro ba 255
img2[img2>255] = 255

img2 = img2.astype('uint8')

plt.title('Image with pixel values increased by 50 and clipped')
plt.imshow(img2)
plt.show()
```
