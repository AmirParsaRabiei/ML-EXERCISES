```python
from sklearn.model_selection  import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'] , random_state=0)


"""
random state hamon seed ma hast ke chon mikhad shuffle kone
chon randomness dare
vaghti in ro taeen konam
sample haye ke random mikone
har bar yeksan khayad bod ba in code

va bara compare 2ta algorithm ham bekar mire
"""


```

```python
"""
inja ham row ha sample ma hast va flower ma va har column ham feature mast 
"""

# create dataframe from data in X_train
# label the columns using the strings in iris_dataset.feature_names
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
iris_dataframe
```

```python
"""
ghabl train kardan inaj ma dar 2d mibinim fazaye kare ma che sheklie

roye ghotre asli chon maana nadare msln
sepal lenght dar barabar sepal lenght pas miad histogramesh ro rasm mikone



hamintor ke inja mibinim 2 gone az flower kheili shabih hastan vali yeki az ona fargh ziadi dare
"""
# create a scatter matrix from the dataframe, color by y_train
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15),
                           marker='o', hist_kwds={'bins': 20}, s=60,
                           alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# inja homan k ke hyperparameter mast to modify mokonim
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
X_train.shape
```

```python
# label mast ke ke kodom 0 hast kodom 1 ya kodom type 2 flower
y_train.shape
```

```python
knn.fit(X_train, y_train)
```

Making predictions:

hala ke ba code bala train kard, roye sample farzi testesh mikonam

```python
"""
inja omadam az ghasd rank2 ijad kardam va ba 4 ta column
ke mishe number of samples and number of features

hala yedone sample mikhay predict koni ham bayad ye matrice 1*4 bedi inja
"""
X_new = np.array([[5,2.9,1,0.2]])
print(X_new.shape)
```

```python
"""
man inja trust nadaram ke dorost train karde ya na


pas bayad beram ro on data haye ke ghabaln nadide testesh konam
"""

prediction = knn.predict(X_new)
print("Prediction:", prediction)
print("Prediction target name: ", iris_dataset['target_names'][prediction])
```

```python
# Evaluating the model

print(X_test.shape)

y_pred = knn.predict(X_test)
print("Test set prediction:\n", y_pred)
```

```python
y_pred == y_test
```

```python
np.sum(y_pred == y_test)
```

```python
len(y_pred)
```

```python
np.round(np.sum(y_pred == y_test)/len(y_pred) * 100, 2)
```

```python
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Test set score: {:.2f}".format(knn.score(X_test,y_test)))
```
